```python
saldo_por_id = car_df.groupby("id")['saldo'].sum().reset_index()
top_10 = saldo_por_id.sort_values(by='saldo', ascending=False).head(10)
```

```python
ids_top_10 = top_10.id.to_list()
```

```python
car_df_top10 = car_df[car_df['id'].isin(ids_top_10)]
car_df_top10.groupby("id")['produto'].size()
```
